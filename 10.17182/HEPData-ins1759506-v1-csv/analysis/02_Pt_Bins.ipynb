# 02 — Transverse Momentum Bins


## Objetivo

Entender os bins de momento transversal.

## Física

O momento transversal é:

$$
p_T = \sqrt{p_x^2 + p_y^2}
$$

Ele mede quanto momento a partícula possui perpendicularmente ao eixo do feixe.

## Regiões físicas

### Soft

Baixo \(p_T\):
- hidrodinâmica;
- bulk do plasma;
- física coletiva.

### Hard

Alto \(p_T\):
- hard probes;
- jets;
- espalhamento perturbativo.


```python

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

DATA_DIR = Path("../")

def load_csv(relative_path):
    return pd.read_csv(DATA_DIR / relative_path, comment="#")

plt.rcParams["figure.figsize"] = (8,5)
```

```python

# Exemplo de leitura e plot

csv_path = "Pb-Pb/Table01.csv"

df = load_csv(csv_path)

print(df.head())

numeric_cols = df.select_dtypes(include=["number"]).columns

if len(numeric_cols) >= 2:
    x = df[numeric_cols[0]]
    y = df[numeric_cols[1]]

    plt.plot(x, y, marker="o")
    plt.xlabel(numeric_cols[0])
    plt.ylabel(numeric_cols[1])
    plt.title(csv_path)
    plt.grid(True)
    plt.show()
```
